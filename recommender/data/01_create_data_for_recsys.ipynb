## Create data for recommender system

```python
import pandas as pd
import json
```

```python
# depth of this file in the project
file_depth = '../..'

with open(file_depth + '/config/data_25m_config.json') as config_file:
    config = json.load(config_file)
```

```python
new_context_path = config['new_context_path']
movies_with_context = pd.read_csv(file_depth + new_context_path + 'movies_with_context.csv')
movies_with_context
```

Keep only movie titles

```python
movie_titles = movies_with_context[['movieId', 'title']]
movie_titles.to_csv('recsys_movie_titles.csv', index=False)
movie_titles
```

Drop title from movie context data

```python
recsys_movies_context = movies_with_context.drop('title', axis=1)
recsys_movies_context.to_csv('recsys_movies_context_data.csv', index=False)
recsys_movies_context
```

Create dataframe from ratings with only userId and movieId

```python
new_context_data = config['new_context_data']
context_data = pd.read_csv(file_depth + new_context_data)
context_data.reset_index(drop=True)
context_data
```

```python
target_data = config['target_data']
target_ratings = pd.read_csv(file_depth + target_data)
target_ratings.reset_index(drop=True)
target_ratings
```

```python
context_data['ratingId'] = context_data.index
context_data = context_data.merge(target_ratings, on='ratingId', how='left')
del target_ratings
context_data = context_data.drop(['ratingId'], axis=1)
context_data
```

Keep only userId and movieId

```python
recsys_ratings_user_movie = context_data[['userId', 'movieId']]
recsys_ratings_user_movie.to_csv('recsys_ratings_only_userid_movieid.csv', index=False)
recsys_ratings_user_movie
```
