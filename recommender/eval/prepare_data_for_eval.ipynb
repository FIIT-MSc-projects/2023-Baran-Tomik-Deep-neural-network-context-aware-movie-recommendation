```python
from sklearn.model_selection import train_test_split
import pandas as pd
import json
```

```python
# depth of this file in the project
file_depth = '../..'

with open(file_depth + '/config/data_25m_config.json') as config_file:
    config = json.load(config_file)
```

### Get suitable users

```python
# added_imdb_context = pd.read_csv(file_depth + '/dataset/ml-1m/new_context/context_data.csv')
added_imdb_context = pd.read_csv(file_depth + '/dataset/ml-25m/new_context/added_imdb_context_with_ratings.csv')
added_imdb_context
```

```python
train, temp = train_test_split(added_imdb_context, test_size=0.2, random_state=42)
del added_imdb_context
val, test = train_test_split(temp, test_size=0.5, random_state=42)
del temp, val
```

```python
test
```

```python
train_user_counts = train['userId'].value_counts()
test_user_counts = test['userId'].value_counts()

# Get userIds of users that have more then 1000 rows in train split and more then 100 rows test split
user_ids = set(train_user_counts.index[train_user_counts >= 3000]) & set(test_user_counts.index[test_user_counts >= 300])

data = []

for user_id in user_ids:
    train_count = train_user_counts.get(user_id, 0)
    test_count = test_user_counts.get(user_id, 0)
    
    data.append({'userId': user_id, 'trainCount': train_count, 'testCount': test_count})
    print(f"User {user_id} (train split: {train_count}, test split: {test_count})")

user_counts_in_splits = pd.DataFrame(data)
```

```python
user_counts_in_splits.to_csv('user_ids_in_train_test_split_25ml.csv', index=False)
test.to_csv('test_split_25ml.csv', index=False)
```
