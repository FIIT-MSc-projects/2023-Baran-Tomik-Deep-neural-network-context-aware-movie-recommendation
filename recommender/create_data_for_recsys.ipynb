```python

# datasets needed to create the final dataset (they are not present in the /recommender repository)
movies_with_context_path = 'data/movies_with_context.csv'
added_imdb_context_path = 'data/added_imdb_context_with_ratings.csv'


data_path = 'data/'
recsys_movies_context_data_path = 'data/recsys_movies_context_data.csv'
```

```python
movies_with_context = pd.read_csv(movies_with_context_path)
movies_with_context
```

Drop title from movie context data

```python
recsys_movies_context = movies_with_context.drop('title', axis=1)
recsys_movies_context.to_csv(data_path + 'recsys_movies_context_data.csv', index=False)
recsys_movies_context
```

Create dataframe from ratings with only userId and movieId

```python
added_imdb_context = pd.read_csv(added_imdb_context_path)
added_imdb_context
```

Keep only userId and movieId

```python
recsys_ratings_user_movie = added_imdb_context[['userId', 'movieId']]
recsys_ratings_user_movie.to_csv(data_path + 'recsys_ratings_only_userid_movieid.csv', index=False)
recsys_ratings_user_movie
```
