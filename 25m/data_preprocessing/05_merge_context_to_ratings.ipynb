# Merge context

```python
import pandas as pd
import json
```

```python
# depth of this file in the project
file_depth = '../..'
```

```python
with open(file_depth + '/config/data_25m_config.json') as config_file:
    config = json.load(config_file)
```

```python
new_context_path = config['new_context_path']
movies_with_context = pd.read_csv(file_depth + new_context_path + 'movies_with_context.csv', encoding="UTF-8")
movies_with_context.reset_index(drop=True)
movies_with_context
```

```python
new_context_ratings = config['new_context_ratings']
ratings = pd.read_csv(file_depth + new_context_ratings, encoding="UTF-8")
ratings.reset_index(drop=True)
ratings
```

```python
rating_context_data = pd.read_csv(file_depth + new_context_path + 'rating_context_data.csv', encoding="UTF-8")
rating_context_data
```

#### Create context data

```python
context_data = ratings.drop(columns=['timestamp'])
context_data = context_data.merge(rating_context_data, on='ratingId', how='left')
context_data = context_data.merge(movies_with_context, on='movieId', how='left')

context_data
```

Check if there are any NaN values

```python
nan_values = context_data[context_data.isna().any(axis=1)]
nan_values
```

There are, 16608 to be precise. That means there were some rows in movies_with_context with no values for matching "movieId" in dataframe ratings.

```python
# for exaplme movieId 720 is not in the movies_with_context (but it is in the ratings)
movies_with_context[movies_with_context["movieId"] == 720]
```

```python
print(f"Assumed number of rows after deletions: {context_data.shape[0]} - {len(nan_values)} = {context_data.shape[0] - len(nan_values)}")
context_data = context_data.dropna()
print('Length of ratings with context after dropping NaN values: ', len(context_data))
```

But we also need to drop movies that were not rated at all.

```python
missing_movieIds = []

for movieId in movies_with_context['movieId']:
    if movieId not in context_data['movieId'].values:
        missing_movieIds.append(movieId)
```

```python
print(f"Number of movies with no rating:  {len(missing_movieIds)}")
print(f"Assumed number of movies after deletions: {movies_with_context.shape[0]} - {len(missing_movieIds)} = {movies_with_context.shape[0] - len(missing_movieIds)}")
movies_with_context = movies_with_context[~movies_with_context['movieId'].isin(missing_movieIds)]
print('Length of movies after dropping movies with no rating: ', len(movies_with_context))
```

```python
movies_with_context.to_csv(file_depth + new_context_path + 'movies_with_context.csv', index=False)
```
