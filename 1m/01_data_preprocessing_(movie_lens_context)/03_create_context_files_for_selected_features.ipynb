```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from joblib import dump
import pandas as pd
import numpy as np
import json
```

```python
# depth of this filr in the project
file_depth = '../..'
path = '../experiment_2_feature_selection/feature_selections_results/'
```

```python
with open(file_depth + '/config/data_1m_config.json') as config_file:
    config = json.load(config_file)

context_data_path = config['context_data']
```

```python
context_data = pd.read_csv(file_depth + context_data_path)
context_data
```

#### Mutual Information

```python
features_selected = []
with open(path + 'mi_selected_features.txt', 'r') as file:
    for line in file:
        features_selected.append(line.strip())

# Dropping columns not present in features_selected
columns_to_drop = [col for col in context_data.columns if col not in features_selected]
mi_context_data = context_data.drop(columns=columns_to_drop)
mi_context_data
```

#### Logistic Regression

```python
features_selected = []
with open(path + 'lr_selected_features.txt', 'r') as file:
    for line in file:
        features_selected.append(line.strip())

# Dropping columns not present in features_selected
columns_to_drop = [col for col in context_data.columns if col not in features_selected]
lr_context_data = context_data.drop(columns=columns_to_drop)
lr_context_data
```

#### L1

```python
features_selected = []
with open(path + 'l1_selected_features.txt', 'r') as file:
    for line in file:
        features_selected.append(line.strip())

# Dropping columns not present in features_selected
columns_to_drop = [col for col in context_data.columns if col not in features_selected]
l1_context_data = context_data.drop(columns=columns_to_drop)
l1_context_data
```

Save context with selected features

```python
feature_selected_path = config['feature_selected_path']

mi_context_data.to_csv(file_depth + feature_selected_path + 'mi_context_data.csv', index=False)
lr_context_data.to_csv(file_depth + feature_selected_path + 'lr_context_data.csv', index=False)
l1_context_data.to_csv(file_depth + feature_selected_path + 'l1_context_data.csv', index=False)
```
