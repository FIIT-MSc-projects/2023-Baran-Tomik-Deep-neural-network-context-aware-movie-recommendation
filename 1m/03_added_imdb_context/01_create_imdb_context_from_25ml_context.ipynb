```python
import pandas as pd
import json
```

```python
# depth of this file in the project
file_depth = '../..'
```

```python
with open(file_depth + '/config/data_1m_config.json') as config_file:
    config = json.load(config_file)
```

```python
new_context_path = config['new_context_path']
movies = pd.read_csv(file_depth + new_context_path + 'movies.csv')
movies
```

```python
movie_ids_1_ml = movies['movie_id'].tolist()
print('Number of movie IDs from 1ml MovieLens dataset: ', len(movie_ids_1_ml))
```

```python
new_context_path = config['new_context_path']
users = pd.read_csv(file_depth + new_context_path + 'users.csv')
users
```

```python
user_ids_1_ml = users['user_id'].tolist()
print('Number of user IDs from 1ml MovieLens dataset: ', len(user_ids_1_ml))
```

```python
context_data_25 = pd.read_csv(file_depth + '/dataset/ml-25m/new_context/context_data.csv')
context_data_25
```

```python
ratings_target_25 = pd.read_csv(file_depth + '/dataset/ml-25m/new_context/ratings_target.csv')
ratings_target_25
```

```python
context_data_25['rating'] = ratings_target_25['rating'].copy()
```

```python
context_data_1 = context_data_25[context_data_25['movieId'].isin(movie_ids_1_ml)]
context_data_1
```

```python
context_data_1 = context_data_1[context_data_1['userId'].isin(user_ids_1_ml)]
context_data_1
```

### Create target ratings

```python
ratings_target = pd.DataFrame()
ratings_target['ratingId'] = context_data_1.index.copy()
ratings_target['rating'] = context_data_1['rating'].copy()
ratings_target
```

```python
context_data_1 = context_data_1.drop(columns=['rating'])
context_data_1
```

```python
dataset_path = config['dataset_path']
context_data_1.to_csv(file_depth + dataset_path + 'added_imdb_context/added_imdb_context.csv', index=False)
ratings_target.to_csv(file_depth + dataset_path + 'added_imdb_context/ratings_target.csv', index=False)
```
